```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the original dataset
data = pd.read_csv('./data/ANC_Data.csv')

# Calculating summary statistics for numerical features
numerical_cols = ['age', 'parity', 'gravidity', 'number_of_antenatal_visits', 'household_income_level', 'gestational_age_at_delivery', 'apgar_score', 'gestational_age_at_first_visit', 'birth_weight_kg']
summary_stats = data[numerical_cols].describe()
print("Summary Statistics:\n", summary_stats)
summary_stats.to_csv('summary_statistics.csv')

# Calculating class imbalance for risk_level
class_counts = data['risk_level'].value_counts()
class_ratios = data['risk_level'].value_counts(normalize=True) * 100
print("Class Counts:\n", class_counts)
print("Class Ratios (%):\n", class_ratios)
class_stats = pd.DataFrame({'Count': class_counts, 'Percentage': class_ratios})
class_stats.to_csv('class_imbalance.csv')

# Plotting histograms for numerical features
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.savefig('numerical_histograms.png')
plt.close()

# Plotting box plots for numerical features by risk_level
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x='risk_level', y=col, data=data)
    plt.title(f'{col} by Risk Level')
plt.tight_layout()
plt.savefig('box_plots_by_risk.png')
plt.close()

# Plotting class distribution of risk_level
plt.figure(figsize=(8, 6))
sns.countplot(x='risk_level', data=data)
plt.title('Class Distribution of Risk Level')
plt.savefig('risk_level_distribution.png')
plt.close()

# Correlation matrix for numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(data[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.savefig('correlation_matrix.png')
plt.close()

print("EDA completed. Plots and statistics saved.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the original dataset
data = pd.read_csv('./data/ANC_Data.csv')

# Define numerical and categorical columns based on the dataset
numerical_cols = ['age', 'parity', 'gravidity', 'number_of_antenatal_visits', 
                  'household_income_level', 'gestational_age_at_delivery', 
                  'apgar_score', 'gestational_age_at_first_visit', 'birth_weight_kg']
categorical_cols = ['education_level', 'marital_status', 'residence', 
                   'birth_complications', 'preterm_birth', 'household_income_level', 
                   'risk_level']

# Calculating summary statistics for numerical features
summary_stats = data[numerical_cols].describe()
print("Summary Statistics:\n", summary_stats)
summary_stats.to_csv('summary_statistics.csv')

# Calculating class imbalance for risk_level
class_counts = data['risk_level'].value_counts()
class_ratios = data['risk_level'].value_counts(normalize=True) * 100
print("Class Counts:\n", class_counts)
print("Class Ratios (%):\n", class_ratios)
class_stats = pd.DataFrame({'Count': class_counts, 'Percentage': class_ratios})
class_stats.to_csv('class_imbalance.csv')

# Plotting histograms for numerical features
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.savefig('numerical_histograms.png')
plt.close()

# Plotting count plots for all categorical features
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=data)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.savefig(f'{col}_distribution.png')
    plt.close()

# Plotting box plots for numerical features by risk_level
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x='risk_level', y=col, data=data)
    plt.title(f'{col} by Risk Level')
plt.tight_layout()
plt.savefig('box_plots_by_risk.png')
plt.close()

# Correlation matrix for numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(data[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.savefig('correlation_matrix.png')
plt.close()

print("EDA completed. Plots and statistics saved.")
```
